```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
width, height, channel = 300,300, 3

my_red_img= np.zeros((width, height, channel), dtype=np.uint8)

my_red_img[:,:,0] = 255
plt.imshow(my_red_img)
plt.axis('off')
plt.title("RED İMAGE")
```

```python
my_green_img= np.zeros((width, height, channel), dtype=np.uint8)
my_green_img[:,:,1] = 255

my_blue_img= np.zeros((width, height, channel), dtype=np.uint8)
my_blue_img[:,:,2] = 255

fig =plt.figure(figsize=(10,7))
rows = 1
columns = 3

fig.add_subplot(rows,columns,1)
plt.imshow(my_red_img, cmap= "gray")
plt.axis('off')
plt.title("RED Image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_green_img, cmap="gray")
plt.axis('off')
plt.title("GREEN Image")

fig.add_subplot(rows,columns,3)
plt.imshow(my_blue_img, cmap="gray")
plt.axis('off')
plt.title("BLUE Image")

```

```python
my_img_white = np.ones((width, height, channel),dtype=np.uint8)*255
my_img_black = np.zeros((width, height, channel),dtype=np.uint8)

my_img_gray_1 = np.random.randint(128,129, (width, height, channel), dtype=np.uint8)
my_img_gray_2 = np.zeros((width, height, channel),dtype=np.uint8)
my_img_gray_2[:,:,:]= 128


fig =plt.figure(figsize=(10,7),facecolor='purple')
rows = 2
columns = 2

fig.add_subplot(rows,columns,1)
plt.imshow(my_img_white)
plt.axis('off')
plt.title("White Image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_img_black)
plt.axis('off')
plt.title("Black Image")

fig.add_subplot(rows,columns,3)
plt.imshow(my_img_gray_1)
plt.axis('off')
plt.title("Gray 1 Image")

fig.add_subplot(rows,columns,4)
plt.imshow(my_img_gray_1)
plt.axis('off')
plt.title("Gray 2 Image")
```
